```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("bank.csv")
import random
```

```python
df.head()
```

```python
job = df['job'].value_counts()
job.describe
```

```python
plt.figure(figsize=(7,4))

plt.bar(job.index,job.values,color='skyblue')
plt.xticks(rotation=45,ha='right')
plt.xlabel('Jobs')
plt.ylabel('Quantity')
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(7,4))

plt.hist(df['age'],bins=30,color='lightgreen',edgecolor='black')
plt.xticks(rotation=45,ha='right')
plt.xlabel('Age')
plt.ylabel('No.of People')
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(7,6))
X=df['duration']
y=df['balance']
plt.scatter(X,y,color='skyblue',alpha=0.5)
plt.xticks(rotation=45,ha='right')
plt.xlabel('Duration')
plt.ylabel('Balance')
plt.tight_layout()
plt.show
```
